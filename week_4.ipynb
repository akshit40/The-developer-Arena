```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('AB_NYC_2019.csv')
df.fillna({'reviews_per_month': 0}, inplace=True)
df.drop(['name', 'host_name', 'last_review'], axis=1, inplace=True)
print("Data loaded and cleaned. Generating plots...\n")
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='neighbourhood_group', data=df, order=df['neighbourhood_group'].value_counts().index, palette='viridis')
plt.title('Number of Airbnb Listings in Each Borough')
plt.xlabel('Number of Listings')
plt.ylabel('Borough')
plt.savefig('w4_listings_per_borough.png')
print("Saved w4_listings_per_borough.png")
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df[df['price'] < 500]['price'], bins=50, kde=True)
plt.title('Distribution of Airbnb Prices (for listings under $500)')
plt.xlabel('Price ($)')
plt.savefig('w4_price_distribution.png')
print("Saved w4_price_distribution.png")
```

```python
plt.figure(figsize=(12, 10))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type', palette='coolwarm', alpha=0.5, s=10)
plt.title('Geographic Distribution of Listings by Room Type')
plt.savefig('w4_geographic_map_by_room_type.png')
print("Saved w4_geographic_map_by_room_type.png")
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df[df['price'] < 500], x='room_type', y='price', palette='pastel')
plt.title('Price Distribution by Room Type (for listings under $500)')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.savefig('w4_price_by_room_type.png')
print("Saved w4_price_by_room_type.png")
```

```python
plt.show() # Optional: displays plots if in an interactive environment
```
