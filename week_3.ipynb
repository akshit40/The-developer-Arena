```python
import pandas as pd

# Load the data into a pandas DataFrame
df = pd.read_csv(FILE_PATH)

print("Data loaded into pandas DataFrame.")
display(df.head())
```

```python
df.fillna({'reviews_per_month': 0}, inplace=True)
df.drop(['name', 'host_name', 'last_review'], axis=1, inplace=True)
print("Data cleaned.\n")
```

```python
borough_stats = df.groupby('neighbourhood_group')['price'].agg(['mean', 'min', 'max'])
print("--- Price Statistics by Borough ---")
print(borough_stats)
print("\n")
```

```python
top_10_reviewed = df.nlargest(10, 'number_of_reviews')
print("--- Top 10 Most Reviewed Listings ---")
print(top_10_reviewed[['neighbourhood', 'room_type', 'price', 'number_of_reviews']])
```
